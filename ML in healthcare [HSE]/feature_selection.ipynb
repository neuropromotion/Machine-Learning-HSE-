## Feature selection is often used to create shorter, faster models. Why do we need several dozen features, if we can achieve a significantly smaller error with a only few? Run your greedy algorithm to find best models with up to 15 features (adding them one by one, as explained).


```python
top_features = []
features_for_cycle = cat_cols + num_cols
top_k_errors = []

for i in range(15):
  errors = []

  for feature in features_for_cycle:
    X = df[top_features + [feature]]
    Y = df[target]

    model = LinearRegression()
    model.fit(X, Y)
    Y_pred = model.predict(X)
    r = Y - Y_pred
    e = abs(r).mean()
    errors.append(e)

  min_ind = errors.index(min(errors))
  top_features.append(features_for_cycle[min_ind])
  features_for_cycle.pop(min_ind)
  top_k_errors.append(min(errors))
```

```python
for i, val in enumerate(top_k_errors):
  print(f'with {i+1} feature(s) error is {val}')
```
