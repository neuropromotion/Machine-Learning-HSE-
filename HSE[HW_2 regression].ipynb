As hydrophobic molecule TG requires support of other molecules, to be transfered within bloodstream. Chylomicrons, Very-low-density lipoprotein (VLDL), Low-density lipoprotein (LDL), High-density lipoprotein (HDL). Those molecules have hydrophylic surface (phosphate group of phospholipids) with proteins (ApoA, ApoB, ApoC, ApoE etc) and fatty nucleus (TGs, cholesterol).

Also ApoE is endogenic ligand to LDL-receptors. Interaction between ApoE and LDL-receptor achive access to consumtion of fat by cells. APOE is located in 19 chromosome near other apolypoprotein's genes. Hightly polymorphic.
* rs174548 - is a single nucleotide polymorphism (SNP) located in the human genome. It represents a specific site in the DNA sequence where one nucleotide can be replaced by another (e.g., adenine can be replaced by guanine or cytosine by thymine). These genetic variations can influence various aspects of health and phenotype. That SNP can cause dysfuction of delta 5 desaturasa enzyme.
* rs4775041 - is another single nucleotide polymorphism (SNP) located in the human genome. Similar to RS174548, it represents a specific location in the DNA sequence where one nucleotide can be substituted for another. SNP were associated either directly with a metabolite range or with a ratio of metabolites. The rarer rs4775041(G) allele was associated with increased phosphatidylethanolamine values.

## Since there are more than 2 categorical features, the contribution of the third and subsequent features will be considered ordinal, even though it is not. This can impede model interpretation and, in general, disrupt mathematical computing.
* Would fields like 'id' cause the model (not necessarily linear) to overfit? Why?
## Fields like 'id' are typically used as unique identifiers for data points and do not contain information that is useful for modeling. Including 'id' fields in the model can lead to overfitting because the model may try to capture the noise or randomness associated with these identifiers. Overfitting occurs when a model learns the training data too well, including the noise, and becomes less effective at making predictions on unseen data. In the case of 'id' fields, they are unlikely to have a meaningful relationship with the target variable, and including them in the model can lead to a loss of generalization ability. Therefore, it is generally advisable to exclude 'id' fields from the modeling process to prevent overfitting.

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
def print_metrics(y_pred, y):
    MSE = round(mean_squared_error(y_pred = y_pred, y_true=y),3)
    RMSE = round(sqrt(MSE),3)
    print(f'R^2: {round(r2_score(y_pred = y_pred, y_true=y),4)}')
    print(f'MSE: {MSE}')
    print(f'RMSE: {RMSE}')
    print(f'MAE: {round(mean_absolute_error(y_pred = y_pred, y_true=y),3)}')
```

```python
X_ = X.copy() # copy our X matrix
X_.reset_index(drop=True, inplace=True) # reset indicies
w0 = np.ones(X_.shape[0]) # create column for w0 weight
X_.insert(loc=0, column='ones', value=w0)
X_
```

```python
import random
np.random.seed(62)
w1 = np.random.randint(0, 300, 19) # create w1 vector of weights
```

```python
def SGD(X, y, w1, max_iter=500, batch_size=10, learning_rate=0.05):
    rows = X.shape[0]
    scores = []
    for iteration in range(max_iter):
        for _ in range(rows // batch_size):

            # Create vector of stochastic numbers
            batch_indices = np.random.choice(rows, batch_size, replace=False)

            # Gain batches and transform to numpy ndarray
            X_batch = (X.iloc[batch_indices]).to_numpy()
            y_batch = (y.iloc[batch_indices]).to_numpy()

            # Compute gradient
            grad = (2 / batch_size) * X_batch.T.dot(X_batch.dot(w1) - y_batch)

            # Update weight's vector and repeat
            w1 = w1 - learning_rate * grad
            scores.append(mean_squared_error(y_pred = X@w1, y_true=y))

    return [w1, scores]
```
